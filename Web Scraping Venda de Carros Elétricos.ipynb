```python
# Importando as Bibliotecas 

import pandas as pd
import requests 
from bs4 import BeautifulSoup
```

```python
# Enviando Requisição para o site 

res = requests.get('https://insideevs.uol.com.br/news/631213/caros-eletricos-mais-vendidos-2022/')

# Verificando se a requisição foi bem sucedida

if res.status_code == 200:
    print("A requisição foi bem-sucedida!")
```

```python
# Realizando o Parser

soup = BeautifulSoup(res.content, 'lxml')

```

```python
# Extraindo o Conteúdo Desejado

table = soup.find_all('table')[0]

print(table)
```

```python
df = pd.read_html(str(table))[0]


```

```python
# Substituindo o valor ausente 

df.fillna(29, inplace=True)
```
